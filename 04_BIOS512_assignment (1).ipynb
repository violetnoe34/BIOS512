```r
library('tidyverse')
```

```r
df_wind = read_csv(
    'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-10-27/wind-turbine.csv')
```

```r
df_wind %>% head
```

Each row in the table represents a single turbine.

Make a bar chart ðŸ“Š **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
df_wind_sorted = df_wind %>%
    mutate(province_territory_sorted = province_territory %>% 
           fct_infreq()%>% 
           fct_rev())

p = ggplot(df_wind_sorted, aes(y = province_territory_sorted))

p = p + geom_bar()

p = p + labs(y = "Province or Territory")

p
```

Make a bar **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
df_lumped = df_wind %>%
    mutate(manufacturer_lumped = manufacturer %>% 
           fct_lump(n = 5) %>% 
           fct_infreq() %>% 
           fct_relevel('Other', after= Inf) %>%
           fct_rev()
    )

p = ggplot(df_lumped, aes(y = manufacturer_lumped))

p = p + geom_bar()

p = p + labs(y = "Manufacturer")

p
```
